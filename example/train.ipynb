```python
import pickle, numpy as np
model = pickle.load(open("model.pkl", "rb"))
```

```python
value = 3
features = np.array([[value]])      # shape: (1, 1)
print("Predicted class for value {}: {}".format(value, model.predict(features)[0]))
prediction = model.predict(features) # e.g. array([1])
print("raw:", prediction, type(prediction), prediction.shape)
print(prediction)
print("class:", int(prediction[0]))
```
